## TASK 2
## Preprocessing Function Explanation

The `preprocess_text()` function is designed to clean and normalize raw text data.

Steps performed:
- Convert text to lowercase for consistency
- Remove URLs and email patterns
- Remove numbers and special characters
- Normalize repeated characters
- Remove extra spaces
- Tokenize text into words
- Filter out short and irrelevant tokens

This ensures the text is ready for machine learning models.

```python
import re

def preprocess_text(text):
    try:
        # Handle empty or invalid input
        if not isinstance(text, str) or text.strip() == "":
            return [], ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'http\S+|www\S+', '', text)

        # Remove emails
        text = re.sub(r'\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Remove emojis and special characters
        text = re.sub(r'[^\w\s]', '', text)

        # Handle repeated characters (soooo → soo)
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenization
        tokens = text.split()

        # Remove short tokens (≤2), but keep "no", "not"
        filtered_tokens = [
            word for word in tokens
            if len(word) > 2 or word in ["no", "not"]
        ]

        cleaned_sentence = " ".join(filtered_tokens)

        return filtered_tokens, cleaned_sentence

    except Exception:
        return [], ""
```

## Task 3
## Stress Testing

The preprocessing function is tested on diverse real-world inputs including:
- Emojis
- Numbers
- URLs
- Repeated characters
- Mixed case text

This ensures the robustness of the pipeline.

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)

    results.append((sentence, tokens, cleaned))
```

## Task 4
## Token Analysis

For each processed sentence, we compute:
- Total number of tokens
- Number of unique tokens
- Average token length

This helps in understanding text complexity and preprocessing effectiveness.

```python
import numpy as np

for sentence, tokens, cleaned in results:
    if len(tokens) == 0:
        continue

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens])

    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {total_tokens}")
    print(f"Unique Tokens: {unique_tokens}")
    print(f"Average Token Length: {avg_length:.2f}")
    print("-" * 50)
```

## Task 5
## Frequency Analysis

We use `Counter` to identify:
- Most frequent words
- Least frequent words

This helps understand common patterns in the dataset.

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

## Task 6
## Full Pipeline

The `full_pipeline()` function processes multiple sentences and returns:
- Combined tokens
- Cleaned sentences

This makes the system scalable and reusable.

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)

print("All Tokens:\n", output["tokens"])
print("\nClean Sentences:\n", output["clean_sentences"])
```

## Task 7
## Error Handling

The system handles edge cases such as:
- Empty input
- Only emojis
- Only numbers

This ensures robustness in real-world scenarios.

```python
edge_cases = [
    "",
    "😍😍😍",
    "123456789"
]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 40)
```
